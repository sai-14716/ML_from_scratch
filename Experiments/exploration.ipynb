```python
import numpy as np
```

```python
#1
l = [i for i in range(1000)]
a = np.array(l, dtype = np.float32).reshape(250, 4)
print(a.shape)
print(a.dtype)

#2
l = [[[i + j + k for k in range(2)] for j in range(3)] for i in range(4)]
b = np.array(l).reshape(6, 4)
print(b.shape)

#3
a = np.linspace(-1, 1, 100).reshape(20, 5)
# print(a)

#4
b = np.arange(0, 30, 1)
b = b.reshape(5, 2, 3)
print(b[1])

#5
a = np.arange(100, 130, 1)
a = a.reshape(3, 2, 5)


#6
b = [a[i].mean() for i in range(3)]
print(b)

b = a[1]
b = [b[i].mean() for i in range(2)]
print(b)


#7
b = a[0].std(axis = 0)
print(b)
b = a[0].std(axis = 1)
print(b)
mean = a[0].mean(axis = 1, keepdims = True)
a[0] = a[0] - mean
print(a[0])

#8 
b = a.max()
print(b)
b = a.argmax()
print(np.unravel_index(b, a.shape))
b = a.min()
print(b)
b = a.argmin()
print(b)

#9
np.where(a < 120, 0, a)

#10
values, counts = np.unique(a[1], return_counts = True)
dic = dict(zip(values, counts))

print(values, counts, dic)
print(dic[114])

#11
print(a[0], a[1])
b = np.concatenate((a[0], a[1]), axis = 0)
print(b)
b = np.hstack((a[0], a[1]))
b = np.vstack((a[0], a[1]))
print(b)

#12
b = np.split(a[0], 5, axis = 1)
print(b)
b = np.vsplit(a[0], 2)
print(b)
b = np.hsplit(a[0], 5)
print(b)

#13
c = np.array(([2, 3], [4, 6]))
print(c.shape)
d = np.tile(c, (3, 3))
print(d)
e = np.repeat(c, 2, axis = 0)
print(e)

#14
a = np.arange(10, 30).reshape(4, 5)
print(a)
b = a > 20
print(b)
b = np.where(a > 15, a, -1)
print(b)

#15 --- Broadcasting
a = np.array(([1, 2, 3], [2, 5, 10]))
b = np.array(([1]))
c = a + b
print(c)

#16
a = np.arange(20).reshape(4, 5)
print(a)
b = a.mean(axis = 0)
print(b)
c = a - b
print(c)
d = a.std(axis = 0)
c = c / d
print(c)

#Inplace Z score
a = (a - a.mean(axis = 0))/a.std(axis = 0)
print(a)

#clipping
a = np.where(a > 1.5, 1.5, a)
a = np.where(a < -1.5, -1.5, a)
print(a)
a_bef = a.copy()
a = np.clip(a, -1.5, 1.5)
b = np.where(a != a_bef, 1, 0)
print(np.sum(b))

#21
percent = (np.sum(b)/np.size(a)) * 100
print(percent)

#22
a = np.array(([10, 5, 3, 8, 9]))
b = np.argsort(a)
print(a, b)

#23
mask = b == 1
c = a[mask]
print(c)

#24
labels = np.array([1, 5, 1, 1, 2, 5, 1])
b = np.bincount(labels)
print(b)

#25
a = np.random.randint(0, 20, (4, 5))
print(a.mean(axis = 0), a.mean(axis = 1))
print(a.std(axis = 0) ** 2, a.std(axis = 1) ** 2)
print(a.var(axis = 0), a.var(axis = 1))
```
